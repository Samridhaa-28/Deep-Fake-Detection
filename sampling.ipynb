```python
import os
import shutil

def clear_folder(folder_path):
    """Delete all files in the destination folder."""
    if os.path.exists(folder_path):
        for filename in os.listdir(folder_path):
            file_path = os.path.join(folder_path, filename)
            try:
                if os.path.isfile(file_path) or os.path.islink(file_path):
                    os.unlink(file_path)  # delete file or symlink
                elif os.path.isdir(file_path):
                    shutil.rmtree(file_path)  # delete subfolder
            except Exception as e:
                print(f"Failed to delete {file_path}. Reason: {e}")
    else:
        os.makedirs(folder_path)  # create folder if it doesn't exist

def copy_files(source_folder, destination_folder, number_of_files):
    """Copy a fixed number of files from source to destination."""
    clear_folder(destination_folder)

    all_files = [f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f))]
    
    files_to_copy = all_files[:number_of_files]

    for file_name in files_to_copy:
        full_source_path = os.path.join(source_folder, file_name)
        full_dest_path = os.path.join(destination_folder, file_name)
        shutil.copy2(full_source_path, full_dest_path)  
        print(f"Copied: {file_name}")

    print(f"\nTotal files copied: {len(files_to_copy)}")

fake_source = r"D:\DL package\venv\frames\fake"
fake_destination = "dataset_small\\fake"
number = 2000  # Number of files to copy

real_source = r"D:\DL package\venv\frames\real"
real_destination = "dataset_small\\real"

copy_files(fake_source, fake_destination, number)
copy_files(real_source, real_destination, number)
```
